```python
# 데이터 타입 조정

변수(칼럼)타입 확인: 데이터.dtypes
- 변수를 범주형으로 변경:
    - 데이터['칼럼명'].astype('category')
- 변수를 수치형으로 변경:
    - 데이터['칼럼명'].astype('int')
    - 데이터['칼럼명'].astype('float')
- NA 값의 처리
    - NA 개수 체크: 데이터.isna().sum()
    - NA값 채우기: 데이터['칼럼명'].fillna(특정숫자)
```

```python
# 라이브러리 사용

import pandas as pd
```

```python
# 1. 과거의 데이터 준비

파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris2.csv'
아이리스 = pd.read_csv(파일경로)
아이리스.head()

# 범주형 데이터(품종 숫자로 표현)
```

```python
# 원핫인코딩

인코딩 = pd.get_dummies(아이리스)
인코딩.head()

# pd.get_dummies() -> 자료들 중 범주형인 것들만 골라 1, 0으로 바꿔주는 코드
# but 이 경우품종에 0,1,2 라고 되어있어서 pandas가 범주형으로 인식 못함
# 따라서 범주형으로 바꿔주는 작업을 해야함
```

```python
print(아이리스.dtypes)
```

```python
# 품종 타입을 범주형으로 바꾸어 준다.
아이리스['품종'] = 아이리스['품종'].astype('category')
print(아이리스.dtypes)
```

```python
# 다시 원핫인코딩

인코딩 = pd.get_dummies(아이리스)
인코딩.head()
```

```python
# NA값을 체크해 본다.

아이리스.isna().sum()
```

```python
아이리스.tail()

# NaN(Not available)이 있으면 인식을 못하므로 숫자로 바꿔주어야 함.
```

```python
# NA값에 꽃잎폭 평균값을 넣어주는 방법

mean = 아이리스['꽃잎폭'].mean()
print(mean)
아이리스['꽃잎폭'] = 아이리스['꽃잎폭'].fillna(mean) #NA값을 mean으로 채워라
아이리스.tail()
```
